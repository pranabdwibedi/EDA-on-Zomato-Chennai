# **To know the top location where the no of hotels are maximum :**

```python
#we have to load all unique location values into another datataset called 'location_unique'
location_unique = Zomato_Chennai['Location'].unique()
#we have defined a blank python list called 'No_Hotel'
No_Hotel = []
for x in location_unique:
  #to add one by one element in the list
  No_Hotel.append(Zomato_Chennai['Location'].value_counts().get(x))
Hotels_Count_Location = pd.DataFrame({'Location' : location_unique, 'No. of Hotels':No_Hotel})
Hotels_Count_Location
```

```python
Top_Location = Hotels_Count_Location.sort_values('No. of Hotels', ascending=False)
Top_10_Location = Top_Location.head(10)
Top_10_Location
sns.barplot(x='Location', y='No. of Hotels', data=Top_10_Location)
```

# **To know the most Common cuisines in Chennai :**

```python
Zomato_Chennai['Cuisine'] = Zomato_Chennai['Cuisine'].apply(eval)
Chennai_Cuisines = Zomato_Chennai.explode('Cuisine')
Chennai_Cuisines.reset_index()
```

```python
#most common cuisines
cuisines_name = Chennai_Cuisines['Cuisine'].unique()
cuisine_count = []
for i in cuisines_name:
   cuisine_count.append(Chennai_Cuisines['Cuisine'].value_counts().get(i))
cuisines = pd.DataFrame({'Cuisines':cuisines_name,'Available in No. of Hotels' : cuisine_count})
cuisines
```

```python
Top_Cuisines = cuisines.sort_values('Available in No. of Hotels',ascending=False)
Top_10_cuisines = Top_Cuisines.head(10)
```

```python
sns.barplot(x='Cuisines', y='Available in No. of Hotels',data = Top_10_cuisines)
```
